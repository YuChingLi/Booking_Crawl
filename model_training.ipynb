```python
from json import load
import pandas as pd
with open('/content/drive/My Drive/booking_crawl.json', 'r', encoding = 'utf-8') as f:
  data = load(f)

data = pd.DataFrame([data['review'], data['status']]).T
data.columns = ['review', 'status']
#data['status'] = data['status'].replace({'good':1, 'bad': 0})
```

```python
from random import sample
#good_df = data[data['status']==1].sample(15000)
#bad_df = data[data['status']==0].sample(15000)
good_df = data[data['status']=='good'].sample(15000)
bad_df = data[data['status']=='bad'].sample(15000)
```

```python
review = pd.concat([good_df,bad_df],ignore_index=True)
#review.shape
review
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
encode_class = le.fit_transform(review['status'])
review['status'] = encode_class
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(review, test_size = 0.3, random_state = 5)
import torch
'''
labels = torch.tensor(train_df['status'].values, dtype=torch.long)
labels
train_df['status'] = labels
'''
train_df
```

```python
from simpletransformers.classification import ClassificationModel
#?ClassificationModel
model = ClassificationModel('bert', 'hfl/chinese-bert-wwm-ext')
model.train_model(train_df, args = {'overwrite_output_dir': True})
```

```python
result, model_output, wrong_predictions = model.eval_model(test_df)
result
```

```python
model.model.save_pretrained('model1')
model.tokenizer.save_pretrained('model1')
model.config.save_pretrained('model1/')
```

```python
import os

# 指定保存路径
save_directory = '/content/drive/MyDrive/outputs'

# 如果目录不存在，则创建该目录
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# 保存模型、分词器和配置文件
model.model.save_pretrained(save_directory)
model.tokenizer.save_pretrained(save_directory)
model.config.save_pretrained(save_directory)
```
